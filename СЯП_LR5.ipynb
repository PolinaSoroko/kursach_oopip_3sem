# **Работа с Pandas.Matplotlib**

2.1 Импортировать датасет. (Нужно взять любой интересующий вас датасет и сделать выборку по параметрам)

2.2 Взять 1000 значений из выбранного датасета

2.3 Проверить данные на пропуски

2.4 Заполнить пропуски и обработать аномальные значения

2.5 Провести полный анализ

2.6 Итоговый обработанный массив без выбросов и пропусков сохраните в файл surname.csv

```python
import pandas as pd
import numpy as np

df=pd.read_csv("test.csv")
print("=== 2.1 ВЫБОРКА ===")
print("   Один столбец (возвращает Series):")
ids = df['Id']
print(f"   Тип: {type(ids)}, первых 5 Id: {list(ids.head(5))}")
print("\n   Несколько столбцов (возвращает DataFrame):")
ids_rooms_kitchen_rooms = df[['Id', 'Rooms', 'KitchenSquare']]
print(f"   Тип: {type(ids_rooms_kitchen_rooms)}, размер: {ids_rooms_kitchen_rooms.shape}")
print(ids_rooms_kitchen_rooms.head(3))
print(f"Первые 10 строк  столбцов Id и Rooms:\n {df.loc[0:10,['Id','Rooms']]}")
print(f"Первые 10 строк и 3 столбца:\n{df.iloc[0:10,0:3]}")
print("\n" + "="*50)
print("=== 2.2 1000 ЗНАЧЕНИЙ ===")
first_1000=df.head(1000)
print(first_1000)
print("\n" + "="*50)
print("=== 2.3 ПРОВЕРКА ДАННЫХ НА ПРОПУСКИ ===")
missing = df[df.isnull().any(axis=1)]
print(f"Строки с пропущенными значениями:\n{missing}")
not_missing = df[df.notnull().all(axis=1)]
print(f"Строки без пропущенных значений:\n{not_missing}")
missing_data = df.isnull().sum()
print(f"\n   Всего пропусков в данных: {df.isnull().sum().sum()}")
print("\n" + "="*50)
print("=== 2.4 Заполнение пропусков и обработка аномальных значений ===")
print("Заполнение медианным значением (LifeSquare):")
LifeSquare_before = df['LifeSquare'].isnull().sum()
median_LifeSquare = df['LifeSquare'].median()
df['LifeSquare'] = df['LifeSquare'].fillna(median_LifeSquare)
LifeSquare_after = df['LifeSquare'].isnull().sum()
print(f"Пропусков в LifeSquare: ДО {LifeSquare_before}, ПОСЛЕ {LifeSquare_after}")
print("\nЗаполнение самым частым значением (Healthcare_1):")
Healthcare_1_before = df['Healthcare_1'].isnull().sum()
most_common_Healthcare_1 = df['Healthcare_1'].mode()[0]
df['Healthcare_1'] = df['Healthcare_1'].fillna(most_common_Healthcare_1)
Healthcare_1_after = df['Healthcare_1'].isnull().sum()
print(f"Пропусков в Healthcare_1: ДО {Healthcare_1_before}, ПОСЛЕ {Healthcare_1_after}")
print("\n" + "="*50)
print("РЕЗУЛЬТАТ ОБРАБОТКИ:")
print(f"   Осталось пропусков: {df.isnull().sum().sum()}")
print("\n" + "="*50)
print("=== 2.5 Проведение полного анализа ===")

print("СОЗДАНИЕ НОВЫХ ПРИЗНАКОВ:")
df['HouseType'] = pd.cut(df['Rooms'],
                       bins=[0,1,2,3,10],
                       labels=['Однокомнатные', 'Двухкомнатные', 'Трехкомнатные', 'Четырехкомнатные'])

print("   Новый столбец создан: HouseType")
print(df[['Id', 'Rooms', 'HouseType']].head())
print("\n" + "="*50)
print("КОМПЛЕКСНЫЙ АНАЛИЗ:")
print("Анализ по годам постройки")
print(df['HouseYear'].describe())
print("\nАнализ площадей по количеству комнат")
rooms_analysis = df.groupby('Rooms').agg({
    'Square': 'mean',
    'LifeSquare': 'mean',
    'KitchenSquare': 'mean',
    'HouseYear': 'mean'
}).round(2)
print(rooms_analysis)
print("\nАнализ этажности по году постройки")
floor_year = df.groupby('HouseYear').agg({
    'HouseFloor': ['mean', 'max', 'count']
}).round(2)
print(floor_year)
print("\nЭкология и характеристики")
eco_analysis = df.groupby('Ecology_3').agg({
    'Square': 'mean',
    'Rooms': 'mean',
    'HouseYear': 'mean'
}).round(2)
print(eco_analysis)

print("\n" + "="*50)
print("ЭКСПОРТ РЕЗУЛЬТАТОВ:")

rooms_df = rooms_analysis.reset_index().add_prefix("rooms_")
floor_df = floor_year.reset_index().add_prefix("floor_")
eco_df = eco_analysis.reset_index().add_prefix("eco_")

combined = pd.concat([rooms_df, floor_df, eco_df], axis=1)
combined.to_csv('Soroko.csv', index=False)

print("\nРезультаты сохранены в 'Soroko.csv'")

print("\n" + "="*50)
print("ИТОГОВЫЕ ВЫВОДЫ:")
print("   На основе проведённого анализа выявлено:")
print("   - Площадь и комфортность жилья увеличиваются с количеством комнат")
print("   - Новые дома имеют большую этажность")
print("   - Экологический фактор демонстрирует связь с типами жилья")
print("   - В некоторых районах наблюдается явная разница по площади и годам постройки")
```

# **Построить графики следующих функций, изменения и диапазон аргумента задать самостоятельно Matplotlib**
Снимок экрана 2025-12-03 004657.png

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
X,Y=np.meshgrid(x,y)
X1=X.copy()
Y1=Y.copy()
X1[X1<0]=0
Z1=X1**0.25+Y1**0.25
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X, Y, Z1)
ax.set_title(r'$z = x^{0.25} + y^{0.25}', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

Z2 = X**2 - Y**2
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z2)
ax.set_title(r'$z = x^2 - y^2$', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

Z3 = 2*X + 3*Y
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z3)
ax.set_title(r'$z = 2x + 3y$', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

Z4 = X**2 + Y**2
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z4)
ax.set_title(r'$z = x^2 + y^2$', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

Z5 = 2 + 2*X + 2*Y - X**2 - Y**2
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z5)
ax.set_title(r'$z = 2 + 2x + 2y - x^2 - y^2$', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
